```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib

import random
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
os.chdir('C:\\Users\\Nevena\\Desktop\\LSDM project\\Data\\train')
lifetime=pd.read_csv('lifetime.csv',header=0,index_col=0,na_values='?')
lifetime_features=pd.read_csv('lifetime_features.csv',header=0,index_col=0,na_values='?')
size=pd.read_csv('size.csv',header=0,index_col=0,na_values='?')
spectrum=pd.read_csv('spectrum.csv',header=0,index_col=0,na_values='?')
labels=pd.read_csv('train_labels.csv',header=0,index_col=0,na_values='?')
scattering=pd.read_csv('scattering.csv',header=0,index_col=0,na_values='?')
```

```python
scattering_np=np.array(scattering)
size_np=np.array(size)
lifetime_np=np.array(lifetime)
lifetime_feat_np=np.array(lifetime_features)
spectrum_np=np.array(spectrum)
#labels_np=np.array(labels)
```

```python
data = np.concatenate((scattering_np, size_np, lifetime_np, lifetime_feat_np, spectrum_np), axis=1)
```

```python
data1 = np.concatenate((scattering_np, lifetime_np, spectrum_np), axis=1)
```

```python
#"Promuckamo" podatke da ne idu redom
indices1 = np.array(range(scattering.shape[0]))

random.seed(0)
random.shuffle(indices1)

new_data = [data[i] for i in indices1]

new_labels = [labels['Pollen'][i] for i in indices1]

new_data1 = [data1[i] for i in indices1]
```

```python
ind = np.int(np.round(0.8 * len(data)))

train = new_data[:ind]
train_labels = new_labels[:ind]
val = new_data[ind:]
val_labels = new_labels[ind:]

train1 = new_data[:ind]
train_labels1 = new_labels[:ind]
```

```python
from sklearn.svm import SVC
from scipy import stats 
```

```python
classification = SVC(kernel='poly')
classification.fit(train, train_labels)
predict = classification.predict(val)
```

```python
cm = confusion_matrix(val_labels, predict)
f1 = f1_score(val_labels, predict, average="weighted")

print('F1 score: ', f1)
print('Confusion matrix: ')
print(cm)
```

```python
print(classification_report(val_labels, predict))
```

```python
classification = SVC(kernel='poly')
classification.fit(train1, train_labels1)
predict1 = classification.predict(val)
```

```python
cm = confusion_matrix(val_labels, predict1)
f1 = f1_score(val_labels, predict1, average="weighted")

print('F1 score: ', f1)
print('Confusion matrix: ')
print(cm)
print(classification_report(val_labels, predict1))
```
